```python
import random

lst = []
for i in range(10_000_000):
    lst.append(random.randint(1, 100))
```

```python
nova_lista = []

for elm in lst:
    if elm % 2 == 0:
        nova_lista.append(elm * 2)

nova_lista.__len__()
```

```python
nova_lista = []

nova_lista = [elm * 2 for elm in lst if elm % 2 == 0]

nova_lista.__len__()
```
